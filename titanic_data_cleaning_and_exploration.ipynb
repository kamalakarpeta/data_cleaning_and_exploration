For this example, let's use the Titanic dataset from seaborn library.

Firstly, we'll import the necessary libraries and load the dataset:

```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = sns.load_dataset('titanic')
```

Next, let's explore the dataset:

```python
# Display the first 5 rows of the dataframe
print(df.head())

# Display the summary statistics of the dataframe
print(df.describe())
```

Now, let's handle missing data. We can use isnull() to check for missing values and sum() to count them:

```python
# Check for missing values
print(df.isnull().sum())
```

For simplicity, we'll fill missing age values with the median age, and drop the 'deck' column as it has many missing values:

```python
# Fill missing age values with median
df['age'].fillna(df['age'].median(), inplace=True)

# Drop the 'deck' column
df = df.drop(['deck'], axis=1)
```

Now let's do some data exploration. We can, for example, visualize the distribution of age and fare, and the survival rate by sex:

```python
# Plot the distribution of 'age' and 'fare'
sns.displot(df['age'])
plt.show()

sns.displot(df['fare'])
plt.show()

# Plot the survival rate by sex
sns.barplot(x='sex', y='survived', data=df)
plt.show()
```

This is a basic example of how you can start with data cleaning and exploration. Depending on the dataset and the specific task, you may need to perform more complex cleaning and exploration steps.
